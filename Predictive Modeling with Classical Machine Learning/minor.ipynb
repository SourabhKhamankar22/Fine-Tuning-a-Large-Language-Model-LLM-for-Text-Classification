### Cell 1: Import Necessary Libraries

First, we need to import the libraries required for data manipulation (pandas), machine learning (scikit-learn), and visualization (matplotlib).

```python
# Import data manipulation and visualization libraries
import pandas as pd
import matplotlib.pyplot as plt

# Import scikit-learn modules for dataset, splitting, training, and evaluation
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

import warnings
warnings.filterwarnings('ignore') # Ignores convergence warnings for cleaner output
```

### Cell 2: Data Acquisition & Preprocessing
We will load the built-in Iris dataset and convert it into a pandas DataFrame. We will also check for missing values to ensure the data is clean.

```python
# Load the Iris dataset
iris = load_iris()

# Create a pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target column to the DataFrame
df['target'] = iris.target

# Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# Display the first 5 rows of the dataset
df.head()
```

### Cell 3: Feature and Target Separation
Next, we need to separate the dataset into our input features ($\mathbf{X}$) and the target variable we want to predict ($\mathbf{y}$).

```python
# Separate features (X) and target (y)
X = df.drop('target', axis=1) # All columns except 'target' are features
y = df['target']              # The 'target' column is what we want to predict

print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")
```

### Cell 4: Train-Test Split
To properly evaluate the model, we must split the data into a Training set (70%) to teach the model, and a Testing set (30%) to test its performance on unseen data.

```python
# Split the data into 70% Training and 30% Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(f"Training data size: {X_train.shape[0]} samples")
print(f"Testing data size: {X_test.shape[0]} samples")
```

### Cell 5: Model Training
Now we will instantiate a Logistic Regression model and train it using our 70% training data.

```python
# Instantiate the Logistic Regression model
# Note: max_iter is increased slightly to help the solver converge
model = LogisticRegression(max_iter=200)

# Train the model using the training data
model.fit(X_train, y_train)

print("Model training complete!")
```

### Cell 6: Evaluation and Metrics
Finally, we will use the trained model to make predictions on the 30% testing set. We will then calculate and report the key performance metrics: Accuracy, Precision, Recall, and the F1-Score. Note: Because the Iris dataset has 3 classes (not just a binary yes/no), we use `average='weighted'` to calculate these scores correctly across all classes.

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)

# Calculate key evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Report the metrics
print("--- Model Evaluation Metrics ---")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")
```

### Model Performance Summary
The Logistic Regression model achieved a perfect score of 1.0 (100%) across all evaluated metrics: Accuracy, Precision, Recall, and F1-Score. 

This indicates that the model was able to flawlessly classify all 45 samples in the unseen testing set into their correct Iris species categories. This perfect performance is common with the Iris dataset because its features (sepal and petal dimensions) provide clear, distinct boundaries between the three different classes of flowers, making it highly suitable for linear classification models.
